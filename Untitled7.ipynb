```python
import pandas as pd

# Replace 'your_dataset.csv' with the exact name of the file you uploaded
# For example: 'grocery_products.csv'
df = pd.read_csv('GroceryDataset.csv')

# Show the first 5 rows of your Kaggle data
print("--- Kaggle Dataset Loaded Successfully ---")
df.head()
```

```python
# 1. See what columns exist and their data types
print("--- Data Info ---")
df.info()

# 2. Check how many missing (null) values are in each column
print("\n--- Missing Values Per Column ---")
print(df.isnull().sum())
```

```python
import numpy as np

# 1. Clean the 'Price' column (remove any strange characters, spaces, or currency symbols, then convert to float)
# We fill the 3 missing prices with the average (mean) price later
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False)
df['Price'] = df['Price'].str.replace(',', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price'] = df['Price'].fillna(df['Price'].mean())

# 2. Handle the 'Rating' column
# Since more than half are missing, we will fill missing ratings with a standard default value like 0.0 or the average rating
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Rating'] = df['Rating'].fillna(0.0)

print("--- Data Cleaning Complete! ---")
# Check the data types again to confirm it worked
df.dtypes
```

```python
import datetime

# 1. Create a intelligent mapping dictionary for different grocery sub-categories
# This sets the standard shelf life (in days) for various types of food items
shelf_life_map = {
    'Meat & Poultry': 3,
    'Dairy & Eggs': 10,
    'Bakery & Desserts': 5,
    'Fresh Produce': 7,
    'Snacks': 90,
    'Beverages': 180,
    'Pantry Staples': 365,
    'Frozen Foods': 120
}

# 2. Map the sub-categories to our dictionary (defaulting to 14 days if the sub-category isn't explicitly listed)
df['Estimated_Shelf_Life_Days'] = df['Sub Category'].map(shelf_life_map).fillna(14).astype(int)

# 3. Simulate a "Purchase Date" of today for these items
today = datetime.date.today()
df['Purchase_Date'] = pd.to_datetime(today)

# 4. Calculate the precise Expiration Date by adding the shelf life days to the purchase date
df['Expiration_Date'] = df['Purchase_Date'] + pd.to_timedelta(df['Estimated_Shelf_Life_Days'], unit='D')

print("--- AI Expiration Date Model Applied Successfully! ---")
# Let's inspect a few rows to see our new smart columns
df[['Title', 'Sub Category', 'Price', 'Expiration_Date']].head(10)
```
